```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import matthews_corrcoef
```

```python
we_df = pd.read_csv('he_she_alignment_just_brands.csv')
we_df['gender'] = we_df['alignment'].apply(lambda x: 'M' if x < 0 else 'F')
```

```python
df_human = pd.read_csv('../Brand-Name-Gender-Prediction/human_names_coded.csv')
df_human = df_human.drop(["dataID"], axis=1)
df_human = df_human.drop(["Name"], axis=1)
x = df_human.drop(['Gender', 'is_unique'], axis=1).as_matrix()[:, 1:]
y = df_human[['Gender']].as_matrix().ravel()
model_LR = LogisticRegression()
model_LR.fit(x, y)
df_brands = pd.read_csv('brand_names_coded_no_duplicates.csv')
df_brands
maps = {}
for i in range(len(df_brands.index)):
    feature_vec = df_brands.iloc[i][4:].as_matrix().ravel().reshape((1, -1))
    x = model_LR.predict_proba(feature_vec)
    print(df_brands.iloc[i][2], x)
    maps[df_brands.iloc[i][2]] = 'M' if x[0][0] > x[0][1] else 'F'
```

```python
keys = list(maps.keys())
vals = list(maps.values())
ss_df = pd.DataFrame({'name': keys, 'gender': vals})
```

```python
web_df = pd.read_csv('brand_website_data_jan.csv', header=1)[['Name', 'Overall Prediction']]
web_df
```

```python
assigned_brands = web_df['Name'].tolist()
we_df_trim = we_df[we_df['name'].isin(assigned_brands)]
ss_df_trim = ss_df[ss_df['name'].isin(assigned_brands)]
```

```python
we_df_trim = we_df_trim.sort_values(by=['name'])
ss_df_trim = ss_df_trim.sort_values(by=['name'])
web_df = web_df.sort_values(by=['Name'])
```

```python
we_df_trim['bin_gen'] = we_df_trim['gender'].apply(lambda x: 1 if x =='M' else 0)
ss_df_trim['bin_gen'] = ss_df_trim['gender'].apply(lambda x: 1 if x =='M' else 0)
web_df['bin_gen'] = web_df['Overall Prediction'].apply(lambda x: 1 if x =='M' else 0)
```

```python
set(assigned_brands).difference(set(we_df_trim['name']))
```

```python
we_ss_corr = matthews_corrcoef(we_df_trim['bin_gen'], ss_df_trim['bin_gen'])
web_we_corr = matthews_corrcoef(web_df['bin_gen'], we_df_trim['bin_gen'])
web_ss_corr = matthews_corrcoef(web_df['bin_gen'], ss_df_trim['bin_gen'])
print("Word Embedding and Sound Symbolism correlation: {}".format(np.round(we_ss_corr, 3)))
print("Word Embedding and Brand Website correlation: {}".format(np.round(web_we_corr, 3)))
print("Sound Symbolism and Brand Website correlation: {}".format(np.round(web_ss_corr, 3)))
```
